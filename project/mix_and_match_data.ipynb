```python
import pandas as pd

# From the audio files recorded we mix and match to get a dataset of true and false positives.
currentDataFile = "data/knockdata.pkl"
currentData = pd.read_pickle(currentDataFile)



validationData = currentData.iloc[4 * 3 : 5 * 3]
validationData = pd.concat([validationData, currentData.iloc[7 * 3 : 8 * 3]])

categories = validationData["label"].unique()

currentData = currentData[~currentData["label"].isin(categories)]

display(currentData)
display(validationData)
```

```python
# Create a new dataframe with four columns: filename1, filename2, label1, label2, match
# filename1 and filename2 are the filenames of the audio files
# label1 and label2 are the labels of the audio files
# match is a boolean value indicating if the labels match
df = pd.DataFrame(columns=["filename1", "filename2", "label1", "label2", "match"])
validationDf = pd.DataFrame(columns=["filename1", "filename2", "label1", "label2", "match"])

# Make a function to fill df will n rows, with p percent of the rows being true positives
def fill_df(df, n, p):
    while(len(df) < n * p):
        # Get a random row from the currentData dataframe
        row = currentData.sample()
        # Get the filename and label of the row
        filename1 = row["filename"].values[0]
        label1 = row["label"].values[0]
        # Get a random row from the currentData dataframe
        row = currentData.sample()
        # Get the filename and label of the row
        filename2 = row["filename"].values[0]
        label2 = row["label"].values[0]
        # If the labels match, then the match value is True
        # Otherwise, the match value is False
        match = label1 == label2

        if(match):
            df.loc[len(df)] = [filename1, filename2, label1, label2, True]

    while(len(df) < n):
        # Get a random row from the currentData dataframe
        row = currentData.sample()
        # Get the filename and label of the row
        filename1 = row["filename"].values[0]
        label1 = row["label"].values[0]
        # Get a random row from the currentData dataframe
        row = currentData.sample()
        # Get the filename and label of the row
        filename2 = row["filename"].values[0]
        label2 = row["label"].values[0]
        # If the labels match, then the match value is True
        # Otherwise, the match value is False
        match = label1 == label2

        if(not match):
            df.loc[len(df)] = [filename1, filename2, label1, label2, match]

    # shuffle the dataframe
    return df.sample(frac=1).reset_index(drop=True)


# Fill the dataframe with 300 rows, with 30% of the rows being true positives
df = fill_df(df, 1000, 0.3)
validDationDf = fill_df(validationDf, 100, 0.3)
```
